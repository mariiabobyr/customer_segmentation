### Data Cleansing

```python
df = df.dropna()
df = df[df["Quantity"] > 0]
df = df[df["UnitPrice"] > 0]
df = df.drop_duplicates()
```

### Introduction of *Revenue*

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
```

### Grouping

```python
customer_df = df.groupby("CustomerID").agg({
    "Revenue" : "sum",
    "InvoiceNo" : "count"
}).rename(columns={
    "Revenue" : "TotalSpent",
    "InvoiceNo" : "TotalOrders"
})

customer_df["AvgOrderValue"] = customer_df["TotalSpent"] / customer_df["TotalOrders"]
```

### Normalization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_df)
```

### Application of k-Means
k-Means-Algorithm is used to form clusters. We use at first Elbow Method so to choose the optimal k-value(number of centroids) for this case.

```python
from sklearn.cluster import KMeans

inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(K_range, inertia)
plt.xlabel("Number of Clusters(k)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

From the graph above we can see, that the optimal k value for this case is 3

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
customer_df["Cluster"] = kmeans.fit_predict(scaled_data)
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
```

### Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

for segment in customer_df["Segment"].unique():
    subset = customer_df[customer_df["Segment"] == segment]
    plt.scatter(subset["TotalSpent"], subset["TotalOrders"], label=segment)

plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    s=300,
    c='red',
    marker='X',
    label='Centroids'
)

plt.xlabel("Total Spent")
plt.ylabel("Total Orders")
plt.title("Customer Segments with Centroids")
plt.legend()
plt.show()
```
